```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Cross Validation scores

```python
# CV results

# Random Forest
mean_RF = 0.0925
std_RF  = 0.0789

# Feed Fordward Neural Network
mean_FFN = 0.0994
std_FFN  = 0.0881

# Recurent Neural Network
mean_RNN = 0.1367
std_RNN  = 0.1044

# Graph Neural Network
mean_GNN = 0.2130
std_GNN  = 0.1150

means = np.array([mean_RF, mean_FFN, mean_RNN, mean_GNN])
stds  = np.array([std_RF, std_FFN, std_RNN, std_GNN])
algorithms = ['RF', 'FFN', 'RNN', 'GNN']
```

```python
# Create point plot with error bars
plt.figure(figsize=(5, 2.5), dpi=300)

# Plot each point separately with its own color
plt.errorbar(algorithms[0], means[0], yerr=stds[0], fmt='v', markersize=8, 
             capsize=3, capthick=1, elinewidth=1.5,
             color='black', markerfacecolor='red',
             markeredgewidth=1.2, label='Random Forest')

plt.errorbar(algorithms[1], means[1], yerr=stds[1], fmt='v', markersize=8, 
             capsize=3, capthick=1, elinewidth=1.5,
             color='black', markerfacecolor='gold',
             markeredgewidth=1.2, label='FFN')

plt.errorbar(algorithms[2], means[2], yerr=stds[2], fmt='v', markersize=8, 
             capsize=3, capthick=1, elinewidth=1.5,
             color='black', markerfacecolor='blue',
             markeredgewidth=1.2, label='RNN')

plt.errorbar(algorithms[3], means[3], yerr=stds[3], fmt='v', markersize=8, 
             capsize=3, capthick=1, elinewidth=1.5,
             color='black', markerfacecolor='green',
             markeredgewidth=1.2, label='GNN')

plt.ylabel('Cross validated error', fontsize=12)
plt.grid(axis='y', alpha=0.3, linestyle='--')

# Add value labels
for i, (mean, std) in enumerate(zip(means, stds)):
    plt.text(i, mean + std + 0.005, f'{mean:.3f}  {std:.3f}', 
             ha='center', va='bottom', fontsize=5)

plt.ylim(-0.1, 0.4)
plt.xlim(-1, 4)

plt.tight_layout()
plt.show()
```

# Adipic acid prediction example

```python
gt_value = np.load('AdipPred-GT-Reactor_3.npy')     # Ground truth
rf_pred  = np.load('AdipPred-RF-Reactor_3.npy')     # Random Forest
ffn_pred = np.load('AdipPred-FFN-Reactor_3.npy')    # Feed Forward Network
rnn_pred = np.load('AdipPred-RNN-Reactor_3.npy')    # Recurrent Neural Network
gnn_pred = np.load('AdipPred-GNN-Reactor_3.npy')    # Graph Neural Network
t = np.linspace(1,8,8)
```

```python
plt.figure(figsize=(8,4), dpi=300)
plt.plot(t, rf_pred, marker='x', markersize=5, label = 'RF', c='red')
plt.plot(t, ffn_pred, marker='x', markersize=5, label = 'FFN', c='gold')
plt.plot(t, rnn_pred, marker='x', markersize=5, label = 'RNN', c='blue')
plt.plot(t, gnn_pred, marker='x', markersize=5, label = 'GNN', c='darkgreen')
plt.plot(t, gt_value, marker='o', markersize=5, label = 'Ground truth', c='black')
plt.xlabel('Time (days)')
plt.ylabel('Adipic Acid (g/L)')
plt.grid(True, linestyle='--', alpha=0.3)
plt.legend();
```
